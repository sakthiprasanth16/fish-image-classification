```python
import os
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping

# To access my data and creating the models folder
data_dir = "data"
train_dir = os.path.join(data_dir, "train")
val_dir = os.path.join(data_dir, "val")
test_dir = os.path.join(data_dir, "test")
save_path = "models"
plot = "model_plots"
os.makedirs(save_path, exist_ok=True)
os.makedirs(plot, exist_ok=True)

# Image input setup parameters
IMG_SIZE = (256, 256)
BATCH_SIZE = 32

# Data augmentation and preprocessing
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    zoom_range=0.2,
    horizontal_flip=True,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    fill_mode='nearest'
)

# Rescaling for validation & test 
val_test_datagen = ImageDataGenerator(rescale=1./255)

# Load data from path
train_gen = train_datagen.flow_from_directory(
    train_dir, target_size=IMG_SIZE, batch_size=BATCH_SIZE, class_mode='categorical', shuffle=True
)

val_gen = val_test_datagen.flow_from_directory(
    val_dir, target_size=IMG_SIZE, batch_size=BATCH_SIZE, class_mode='categorical', shuffle=False
)

test_gen = val_test_datagen.flow_from_directory(
    test_dir, target_size=IMG_SIZE, batch_size=BATCH_SIZE, class_mode='categorical', shuffle=False
)

# Custom CNN model architecture 
def create_custom_cnn(input_shape, num_classes):
    model = Sequential([
        Conv2D(32, (3, 3), activation='relu', input_shape=input_shape), # 3 conv layer and 3 pooling layer 
        MaxPooling2D(2, 2),
        Conv2D(64, (3, 3), activation='relu'),
        MaxPooling2D(2, 2),
        Conv2D(128, (3, 3), activation='relu'),
        MaxPooling2D(2, 2),
        Flatten(),
        Dropout(0.5), # Reduce overfitting
        Dense(128, activation='relu'),
        Dense(num_classes, activation='softmax') # Softmax layer for multi-class output 
    ])
    return model

def plot_training(history, name):
    plt.figure(figsize=(10, 4))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train')
    plt.plot(history.history['val_accuracy'], label='Val')
    plt.title(f"{name} Accuracy")
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train')
    plt.plot(history.history['val_loss'], label='Val')
    plt.title(f"{name} Loss")
    plt.legend()
    plt.tight_layout()
    plt.show()
    
    plt.tight_layout()
    plt.savefig(os.path.join(plot, f"plot_{name}.png"))
    plt.close()

# Calling the function and compiling
model = create_custom_cnn((256, 256, 3), train_gen.num_classes) # In RGB
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Callbacks for monitor to save a improved model
# Saves the best model based on validation accuracy
checkpoint = ModelCheckpoint(
    filepath=os.path.join(save_path, "custom_cnn.h5"),
    monitor='val_accuracy',
    save_best_only=True,
    mode='max',
    verbose=1
)
# To avoid overfitting, stops training early if validation accuracy doesnâ€™t improve for 3 epochs
early_stop = EarlyStopping(monitor='val_accuracy', patience=3)

# Training the model with 10 epochs
history = model.fit(train_gen, validation_data=val_gen, epochs=10, callbacks=[checkpoint, early_stop])

name = "CNN"
plot_training(history, name)
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report
from tensorflow.keras.models import Model
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
from tensorflow.keras.applications import VGG16, ResNet50, MobileNet, InceptionV3, EfficientNetB0

# To access my data, creating the model and plot folder 
data_dir = "data"  # make sure to match your folder structure
train_dir = os.path.join(data_dir, "train")
val_dir = os.path.join(data_dir, "val")
test_dir = os.path.join(data_dir, "test")
save_dir = "models"
plot = "model_plots"
os.makedirs(save_dir, exist_ok=True)
os.makedirs(plot, exist_ok=True)


# Image input setup parameters and max epochs
IMG_SIZE = (256, 256)
BATCH_SIZE = 32
EPOCHS = 10  
```

```python
# Data augmentation and preprocessing
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    zoom_range=0.2,
    horizontal_flip=True,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.2,
    fill_mode='nearest'
)

# Rescaling for validation & test
val_test_datagen = ImageDataGenerator(rescale=1./255)

# Load data from path
train_gen = train_datagen.flow_from_directory(
    train_dir, target_size=IMG_SIZE, batch_size=BATCH_SIZE, class_mode='categorical', shuffle=True
)

val_gen = val_test_datagen.flow_from_directory(
    val_dir, target_size=IMG_SIZE, batch_size=BATCH_SIZE, class_mode='categorical', shuffle=False
)

test_gen = val_test_datagen.flow_from_directory(
    test_dir, target_size=IMG_SIZE, batch_size=BATCH_SIZE, class_mode='categorical', shuffle=False
)
```

```python
# Building the final model
def create_model(base_model, num_classes):
    base_model.trainable = False
    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = Dropout(0.5)(x)
    x = Dense(128, activation='relu')(x)
    output = Dense(num_classes, activation='softmax')(x)
    return Model(inputs=base_model.input, outputs=output)

# Visualizing model performance saving as png
def plot_training(history, name):
    plt.figure(figsize=(10, 4))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train')
    plt.plot(history.history['val_accuracy'], label='Val')
    plt.title(f"{name} Accuracy")
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train')
    plt.plot(history.history['val_loss'], label='Val')
    plt.title(f"{name} Loss")
    plt.legend()
    plt.tight_layout()
    plt.show()
    
    plt.tight_layout()
    plt.savefig(os.path.join(save_dir, f"plot_{name}.png"))
    plt.close()

# This function train and save the model
def run_model(base_model,model_name):
    model = create_model(base_model, train_gen.num_classes)
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

    # Callbacks for monitor to save a improved model
    # Saves the best model based on validation accuracy
    checkpoint_cb = ModelCheckpoint(
        filepath=os.path.join(save_dir, f"best_{model_name}.h5"),
        monitor='val_accuracy',
        save_best_only=True,
        mode='max',
        verbose=1
    )
    earlystop_cb = EarlyStopping(
        monitor='val_accuracy',
        patience=3,
        restore_best_weights=True,
        verbose=1
    )

    # Training history notes
    history = model.fit(
        train_gen,
        validation_data=val_gen,
        epochs=EPOCHS,
        callbacks=[checkpoint_cb, earlystop_cb]
    )

    plot_training(history, model_name)
```

```python
# This for VGG16 model
model_name = "VGG16"
print(f"ğŸš€ Training: {model_name}")

base_model = VGG16(weights='imagenet', include_top=False, input_shape=(256, 256, 3))

# Here calling the function
run_model(base_model,model_name)
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Paths
data_dir = "data"
test_dir = os.path.join(data_dir, "test")
model_path = "models/custom_cnn.h5"
history_path = "models/history.npy"

IMG_SIZE = (256, 256)
BATCH_SIZE = 32

# Load test data
test_datagen = ImageDataGenerator(rescale=1./255)

test_gen = test_datagen.flow_from_directory(
    test_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)

# Load trained model
model = load_model(model_path)

# Predict
y_true = test_gen.classes
class_names = list(test_gen.class_indices.keys())

y_pred = model.predict(test_gen)
y_pred_classes = np.argmax(y_pred, axis=1)

# Classification Report
print("ğŸ“ Classification Report:\n")
print(classification_report(y_true, y_pred_classes, target_names=class_names))

# Confusion Matrix
cm = confusion_matrix(y_true, y_pred_classes)
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=class_names, yticklabels=class_names, cmap='Blues')
plt.title("Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.tight_layout()
plt.show()

# Plot training history
history = np.load(history_path, allow_pickle=True).item()

def plot_training(history):
    plt.figure(figsize=(12, 5))

    # Accuracy
    plt.subplot(1, 2, 1)
    plt.plot(history['accuracy'], label='Train Accuracy')
    plt.plot(history['val_accuracy'], label='Validation Accuracy')
    plt.title("Model Accuracy")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.legend()

    # Loss
    plt.subplot(1, 2, 2)
    plt.plot(history['loss'], label='Train Loss')
    plt.plot(history['val_loss'], label='Validation Loss')
    plt.title("Model Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend()

    plt.tight_layout()
    plt.show()

plot_training(history)
```
